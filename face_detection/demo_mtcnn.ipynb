```python
# 导入包
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi']=120
```

```python
# 读取照片
img = cv2.imread('./images/faces2.jpg')
# MTCNN需要RGB通道顺序
img_cvt = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_cvt)  #kernel appear to have failed 
```

```python
# 导入MTCNN
# conda install -c conda-forge mtcnn
# conda install -c conda-forge tensorflow
from mtcnn.mtcnn import MTCNN
```

```python
# 加载模型
mtcnn_face_detector = MTCNN()
#检测人脸
detections = mtcnn_face_detector.detect_faces(img_cvt)
detections
```

```python
img_cvt_copy = img_cvt.copy()
#解析结果
for face in detections:
    if face['confidence'] > 0.9:
        x,y,w,h = face['box']
        print((x,y,w,h),face['confidence'])
        cv2.rectangle(img_cvt_copy,(x,y),(x+w,y+h),(0,255,0),5)
plt.imshow(img_cvt_copy)
```
