```python
# 导入包
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi']=120
```

```python
img = cv2.imread('./images/faces2.jpg')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
# 加载模型
ssd_face_detector = cv2.dnn.readNetFromCaffe('./weights/deploy.prototxt.txt','./weights/res10_300x300_ssd_iter_140000.caffemodel')
img.shape
```

```python
# 原图尺寸
img_h = img.shape[0]
img_w = img.shape[1]
#缩放
img_resize = cv2.resize(img,(500,300))
#图像转为blob
img_blob = cv2.dnn.blobFromImage(img_resize,1,(500,300),(104.0, 177.0, 123.0))
```

```python
#输入
ssd_face_detector.setInput(img_blob)
```

```python
#推理
detections = ssd_face_detector.forward()
detections
```

```python
detections.shape
```

```python
# 检测的人脸数量
num_of_detections = detections.shape[2]
num_of_detections
```

```python
img_copy = img.copy()
for index in range(num_of_detections):
    # 置信度
    detection_confidence = detections[0,0,index,2]
    if detection_confidence > 0.3:
        print(detections[0,0,index,3:7])
        locations = detections[0,0,index,3:7] * np.array([img_w,img_h,img_w,img_h])
        # print(locations)
        
        lx,ly,rx,ry = locations.astype(np.int32)
        # print((lx,ly,rx,ry),detection_confidence)
        
        cv2.rectangle(img_copy,(lx,ly),(rx,ry),(0,255,0),3)
plt.imshow(cv2.cvtColor(img_copy,cv2.COLOR_BGR2RGB))
```
