```python
import cv2
import dlib
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline
plt.rcParams['figure.dpi'] = 120
```

```python
img = cv2.imread('./images/faces1.jpg')
#构造人脸检测器
hog_face_detector = dlib.get_frontal_face_detector()
#检测人脸
detections = hog_face_detector(img,1)
type(detections)
```

```python
detections[0]
```

```python
# left----->矩形区域左上角的横坐标
# top------>矩形区域左上角的纵坐标
# right---->矩形区域右下角的横坐标
# down----->矩形区域右上角的纵坐标
# rectangle类的构造函数接受四个参数，全部为int型。
```

```python
#解析人脸检测结果
for face in detections:
    x = face.left()
    y = face.top()
    w = face.right()
    h = face.bottom()
    print((x,y,w,h))
    cv2.rectangle(img,(x,y),(w,h),(0,255,0),3)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
img = cv2.imread('./images/faces2.jpg')
#构造人脸检测器
hog_face_detector = dlib.get_frontal_face_detector()
#检测人脸
detections = hog_face_detector(img,1)
#解析人脸检测结果
for face in detections:
    x = face.left()
    y = face.top()
    w = face.right()
    h = face.bottom()
    cv2.rectangle(img,(x,y),(w,h),(0,255,0),3)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
