```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline
plt.rcParams['figure.dpi'] = 120
```

```python
img = cv2.imread('./images/faces1.jpg')
img_fixed = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_fixed)
```

```python
#构造cascade级联检测器
face_detector = cv2.CascadeClassifier('./cascades/haarcascade_frontalface_default.xml')
#灰度图
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#检测结果
detections = face_detector.detectMultiScale(img_gray)

detections
```

```python
for (x,y,w,h) in detections:
    # print((x,y,w,h))
    cv2.rectangle(img_fixed,(x,y),(x+w,y+h),(0,255,0),3)
plt.imshow(img_fixed)
```

```python
img = cv2.imread('./images/faces1.jpg')
img_fixed = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#构造cascade级联检测器
face_detector = cv2.CascadeClassifier('./cascades/haarcascade_frontalface_default.xml')
#灰度图
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#检测结果
detections = face_detector.detectMultiScale(img_gray,scaleFactor=1.21,minNeighbors=6)
for (x,y,w,h) in detections:
    # print((x,y,w,h))
    cv2.rectangle(img_fixed,(x,y),(x+w,y+h),(0,255,0),3)
plt.imshow(img_fixed)
```

```python
img = cv2.imread('./images/faces2.jpg')
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
detections = face_detector.detectMultiScale(img_gray,scaleFactor=1.24,minSize=(40,40))

for (x,y,w,h) in detections:
    print((w,h))
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
