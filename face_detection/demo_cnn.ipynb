```python
import cv2
import dlib
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline
plt.rcParams['figure.dpi'] = 120
```

```python
img = cv2.imread('./images/faces1.jpg')
#构造人脸检测器,传入预训练模型./weights/mmod_human_face_detector.dat
cnn_face_detector = dlib.cnn_face_detection_model_v1('./weights/mmod_human_face_detector.dat')
gray = cv2.cvtColor(img.copy(),cv2.COLOR_BGR2GRAY)
```

```python
#人脸检测
detections = cnn_face_detector(gray,2)
detections
```

```python
detections[0].rect
```

```python
len(detections)
```

```python
for face in detections:
    x = face.rect.left()
    y = face.rect.top()
    r = face.rect.right()
    b = face.rect.bottom()
    #confidence 置信度
    print((x,y,r,b),face.confidence) 
    cv2.rectangle(img,(x,y),(r,b),(0,255,0),3)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
