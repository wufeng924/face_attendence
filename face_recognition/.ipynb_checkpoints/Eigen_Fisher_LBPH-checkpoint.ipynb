```python
#构造hog人脸检测器
hog_face_detector = dlib.get_frontal_face_detector()
```

```python
#获取图片文件中图片人脸区域和label标签号
def getFileImgLable(fileName):
    cap = cv2.VideoCapture(fileName)
    ret,img = cap.read()
    
    #转为灰度图
    img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    
    detections = hog_face_detector(img,1)
    if len(detections)>0:
        x = detections[0].left()
        y = detections[0].top()
        w = detections[0].right()
        h = detections[0].bottom()
        #获取人脸区域
        img_crop = img[y:h,x:w]
        #统一缩放 解决冲突 Eigenface、Fisherface 
        img_crop = cv2.resize(img_crop,(120,120))
        #获取人脸lable id
        lable_id = int(fileName.split('/')[-1].split('.')[0].split('subject')[-1])

        return img_crop,lable_id
    else:
        return None,-1
```

```python
import glob
```

```python
file_list = glob.glob('./yalefaces/train/*')
len(file_list)
```

```python
img_list = []
lable_list = []
for train_file in file_list:
    img,lable = getFileImgLable(train_file)
    #过滤数据
    if lable != -1:
        img_list.append(img)
        lable_list.append(lable)
```

```python
len(img_list)
```

```python
len(lable_list)
```

```python
img_list[0]
```

```python
img_list[0].shape
```

```python
#构造分类器
face_cls = cv2.face.LBPHFaceRecognizer_create()
# face_cls = cv2.face.EigenFaceRecognizer_create()
# face_cls = cv2.face.FisherFaceRecognizer_create()
```

```python
#训练
face_cls.train(img_list,np.array(lable_list))
```

```python
test_file = './yalefaces/test/subject09.sad.gif'
img,lable = getFileImgLable(test_file)

if lable != -1:
    predict_label,confidence = face_cls.predict(img)
    print(predict_label,confidence)
```

```python
#评估模型
file_list = glob.glob('./yalefaces/test/*')

true_list = []
pre_list = []

for test_file in file_list:
    img,lable = getFileImgLable(test_file)
    
    if lable != -1:
        predict_label,confidence = face_cls.predict(img)
        pre_list.append(predict_label)
        true_list.append(lable)
```

```python
#查看准确率
from sklearn.metrics import accuracy_score
accuracy_score(true_list,pre_list)
```

```python
#融合矩阵
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(true_list,pre_list)
cm
```

```python
#可视化
import seaborn
plt.rcParams['figure.dpi'] = 80
seaborn.heatmap(cm,annot=True)
```

```python
#保存模型
face_cls.save('./weights/Eigen.yml')
```

```python
test_file = './yalefaces/test/subject14.sad.gif'
img,lable = getFileImgLable(test_file)

if lable != -1:
    predict_label,confidence = new_cls.predict(img)
    print(predict_label,confidence)
```
